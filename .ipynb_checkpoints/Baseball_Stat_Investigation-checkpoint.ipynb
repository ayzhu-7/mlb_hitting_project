```python
def spray_angle(hc_x, hc_y, stand):
    mult = 1
    if stand == 'L':
        mult = -1
    return mult * np.degrees(np.arctan2((hc_x - 125.42), (198.27 - hc_y)))
```

```python
phillies_roster = statsapi.roster(143)
# print(phillies_roster)
# print([name.split(' ') for name in phillies_roster.split('\n')])
phillies_hitters = [' '.join(name.split(' ')[-2:]) for name in phillies_roster.split('\n') if len(name)>2 and name.split(' ')[2]!='P']
phillies_hitters = [name.split(' ') for name in phillies_hitters]
phillies_hitters
```

```python
phillies_ids = [playerid_lookup(name[1].lower(), name[0].lower()) for name in phillies_hitters if name[0] != 'J.T.'] + [playerid_lookup('realmuto')]
phillies_id_numbers = [player['key_mlbam'].iloc[0] for player in phillies_ids]
```
